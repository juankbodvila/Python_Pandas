```python
import pandas as pd
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
nba2022 = pd.read_html(url,attrs={'id':'per_game_stats'})
nba2022 = nba2022[0]
nba2022.index.rename('Index',inplace=True)
nba2022 = nba2022.loc[(nba2022.Rk!="Rk")]
nba2022
```

```python
nba2022.sort_values('PTS',ascending=False)
```

```python
pd.set_option('display.max_columns', None)
```

```python
nba2022.loc[nba2022['Player']=="Stephen Curry"]
```

```python
#Looking for the oldest players
nba2022.sort_values('Age',ascending=False).head(5)
```

```python
#Looking for the best 3 point shooters
nba2022.sort_values('3P',ascending=False).head(5)
```

```python
#Looking for the best Free Throw shooter
#careful, some of them are not eligible. They only made 1 FT and nailed it...
nba2022.sort_values('FT%',ascending=False).head(5)
```

```python
#Most games in a season
#Careful, the original DataFrame did not have the column "G" as float, so we need to convert it in order to sort
nba2022["G"] = nba2022["G"].astype(float)
nba2022.sort_values('G',ascending=False)
```

```python
#most turnovers per game
nba2022.sort_values('TOV',ascending=False)
```

```python
#most blocks per game
nba2022.sort_values('BLK',ascending=False).head(5)
```

```python
#most assists per game
nba2022['AST'] = nba2022['AST'].astype(float)
nba2022.sort_values('AST',ascending=False).head(5)
```

```python
# The Dikembe Mutombo Prize:
nba2022.sort_values('BLK',ascending=False).head(5)
```

```python
# The Dennis Rodman prize:
nba2022["DRB"] = nba2022["DRB"].astype(float)
nba2022.sort_values('DRB',ascending=False).head(3)
```
