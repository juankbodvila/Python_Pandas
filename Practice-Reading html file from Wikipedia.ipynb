```python
import pandas as pd
```

```python
# Importing wikipedia page on world population
url = 'http://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)#'
worldpop = pd.read_html(url)
df = worldpop[0] #we select the 1st table in the website
df = df.rename(columns={"Country/Area":"Country","UN continentalregion[4]":"Continent"})
df.drop("UN statisticalsubregion[4]",inplace=True,axis=1)
df
```
